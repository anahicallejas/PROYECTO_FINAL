```python
# librerias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
```

```python
# carga dataset
file_path = './DATA/issues_history_eda_fe.csv'
data = pd.read_csv(file_path, encoding='utf-8')
data
```

```python
# Ver la cantidad de valores faltantes por columna
missing_values = data.isnull().sum()
missing_values_table = pd.DataFrame(missing_values, columns=["Missing Values"])
missing_values_table
```

```python
missing_values = data.isnull().sum()
missing_values = missing_values[missing_values > 0]
print(missing_values)
```

```python
data['Resolution Time'].fillna(data['Resolution Time'].median(), inplace=True)
```

```python
data.info()
```

```python
data['Story Points'] = data['Story Points'].astype('category')
data['Priority'] = data['Priority'].astype('category')
data['Parent Key'] = data['Parent Key'].astype('category')
data['Assignee'] = data['Assignee'].astype('category')
data['Key'] = data['Key'].astype('category')
data['Issue Type'] = data['Issue Type'].astype('category')
data['Sprint'] = data['Sprint'].astype('category')
data['Team'] = data['Team'].astype('category')
data['Blocked'] = data['Blocked'].astype('category')
data['Complexity'] = data['Complexity'].astype('category')
```

```python
from sklearn.preprocessing import MinMaxScaler

# Inicializar el escalador
scaler = MinMaxScaler()

# Escalar todas las columnas numéricas automáticamente
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
numeric_columns = [col for col in numeric_columns if col != 'Resolution Time']
data[numeric_columns] = scaler.fit_transform(data[numeric_columns])
data
```

```python
# one hot encoding
# Seleccionar columnas categóricas
columns_to_exclude = ['Key', 'Blocked', 'Estimation Accurate']
category_columns = [col for col in data.select_dtypes(include=['category']).columns if col not in columns_to_exclude]

# Aplicar pd.get_dummies solo a las columnas categóricas seleccionadas
data = pd.get_dummies(data, columns=category_columns, dtype=int, drop_first=True)
data
```

```python
data.dtypes
```

```python
output_csv_path = './DATA/issues_history_ready.csv'
data.to_csv(output_csv_path, index=False)

output_csv_path
```
